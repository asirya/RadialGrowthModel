```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d
```

```python
# Define points
points = [[0.2, 0.2], [0.8, 0.2],[0.5, 0.7]]

points = np.append(points, [[999,999], [-999,999], [999,-999], [-999,-999]], axis = 0)  # For colorizing

vor = Voronoi(points)

# Plot the Voronoi diagram
fig, ax = plt.subplots(figsize=(6, 6))

# Define colors for regions
colors = ['lightblue', 'lightgreen', 'lightcoral']

# Loop through regions and color them
for idx, region_index in enumerate(vor.point_region):
    region = vor.regions[region_index]
    if -1 not in region and len(region) > 0:  # Check for finite regions
        polygon = [vor.vertices[i] for i in region]
        ax.fill(*zip(*polygon), color=colors[idx % len(colors)], alpha=0.6, edgecolor='black')
        
voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='white', line_width=3, point_size=10)
# Plot the points
ax.scatter(*zip(*points), color='red', s=100)

# Formatting the plot
ax.set_xlim(-0.5, 1.5)
ax.set_ylim(-0.5, 1.5)
ax.set_aspect('equal', adjustable='box')
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```
